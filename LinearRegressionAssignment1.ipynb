```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("salary.csv")
x=np.array(dataset[['x']])
y=np.array(dataset[['y']])
x,y
```

```python
plt.scatter(x,y)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
print("Mean for x value",x_mean,",","Mean for y values",y_mean)
```

```python
#to calculate value of cofiicient and intercept
num=0
den=0
for i in range(len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)**2
    b1=num/den
    b0=y_mean-(b1*x_mean)
print("intercept",b0,"cofficient",b1)
```

```python
y_pred=b0+b1*x
y_pred
```

```python
plt.scatter(x,y,c="red")
plt.plot(x,y_pred)
plt.show()
```

```python
#R-square=1-(ssr/sst)
num_r=0
den_r=0
for i in range(len(x)):
    num_r+=(y[i]-y_pred[i])**2
    den_r+=(y[i]-y_mean)**2
    r_square=1-num_r/den_r
print("The value of coefficient of determination R_sqaure",r_square)
```
